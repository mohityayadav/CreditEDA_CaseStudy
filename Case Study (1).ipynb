## Binning the age groups to check the trend of defaulters

```python

#Age information into a separate dataframe


age_data = data2[['TARGET', 'DAYS_BIRTH']]

age_data['YEARS_BIRTH'] = (age_data['DAYS_BIRTH'] / 365)


# Bin the age data
age_data['YEARS_BINNED'] = pd.cut(age_data['YEARS_BIRTH'], bins = np.linspace(20, 70, num = 11))
age_data.head(10)
```

```python
groups  = age_data.groupby('YEARS_BINNED').mean()
groups
```

```python
plt.figure(figsize = (7, 8))

# Graph the age bins and the average of the target as a bar plot
plt.bar(groups.index.astype(str), 100 * groups['TARGET'])

# Plot labeling
plt.xticks(rotation = 45); plt.xlabel('Age Group (years)'); plt.ylabel('Defaulters (%)')
plt.title('Defaulters by Age Group');
```
